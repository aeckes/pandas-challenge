```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas Data Frames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
```

## District Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average math score 

* Calculate the average reading score

* Calculate the overall passing rate (overall average score), i.e. (avg. math score + avg. reading score)/2

* Calculate the percentage of students with a passing math score (70 or greater)

* Calculate the percentage of students with a passing reading score (70 or greater)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
school_data_complete.columns
```

```python
#Calculate total number of schools

school_data_complete['type'].str.strip('[]')

total_schools = len(school_data_complete['school_name'].unique())

school_types = school_data_complete['type'].unique()

#print(school_types)

total_students = school_data_complete['student_name'].count()

total_budget = sum(school_data['budget'])

avg_math_score = sum(school_data_complete['math_score']) / total_students
avg_reading_score = sum(school_data_complete['reading_score']) / total_students

ovrall_passing_rate = (avg_math_score+avg_reading_score)/2

students_70plus_math = school_data_complete.loc[school_data_complete['math_score'] >= 70, ['math_score']].count()/total_students
students_70plus_reading = school_data_complete.loc[school_data_complete['reading_score'] >= 70 , ['reading_score']].count()/total_students

#convert single row dataframe to single value
students_70plus_math = students_70plus_math.iat[0]
students_70plus_reading = students_70plus_reading.iat[0]

```

```python
#put results in dataframe

summary_df = pd.DataFrame([[total_schools,
        total_students,
        total_budget,
        avg_math_score,
        avg_reading_score,
        ovrall_passing_rate,
        students_70plus_math,
        students_70plus_reading]],columns=["Total Schools",
        "Total Students",
        "Total Budget",
        "Avg Math Score",
        "Avg Reading Score",
        "Overall Passing Rate",
        "% Passing Math",
        "% Passing Reading"])

summary_df
```

## Top Performing Schools (By Passing Rate)

* Sort and display the top five schools in overall passing rate

```python
school_grp = school_data_complete.groupby(['school_name'])
#school_grp.count().head(10)

school_names = school_grp['school_name']

#not preferred, causes issue w/ last piece of activity (show results by school type), can't figure out alternative
#school_types = school_grp['type'].unique()

total_students_grp = school_grp['student_name'].count()
#print(total_students_grp)

total_school_budget = school_grp['budget'].sum()/total_students_grp
#print(total_school_budget)

per_student_budget = total_school_budget/total_students_grp
#print(per_student_budget)

avg_math_score_grp = school_grp['math_score'].mean()
#print(avg_math_grp)

avg_reading_score_grp = school_grp['reading_score'].mean()
#print(avg_reading_score_grp)

passing_math = school_data_complete[school_data_complete['math_score'] >=70].groupby('school_name')['math_score'].count()/total_students_grp
#print(passing_math_count)

passing_reading = school_data_complete[school_data_complete['reading_score'] >=70].groupby('school_name')['reading_score'].count()/total_students_grp
#print(passing_reading)

overall_passing = (passing_math + passing_reading) / 2
#print(overall_passing)

```

```python
#add to dataframe

school_results_data = {"School Type": school_types,"Total Students":total_students_grp, "Total School Budget":total_school_budget, "Per Student Budget":per_student_budget,
                       "Average Math Score":avg_math_score_grp,"Average Reading Score":avg_reading_score_grp, "% Passing Math":passing_math, "% Passing Reading":passing_reading, "% Overall Passing Rate":overall_passing}

df_school_results = pd.DataFrame.from_dict(school_results_data)

df_top5 = df_school_results.sort_values('% Overall Passing Rate', ascending=False)

df_top5.head(5)
```

## Math Scores by Grade

* Create a table that lists the average Reading Score for students of each grade level (9th, 10th, 11th, 12th) at each school.

  * Create a pandas series for each grade. Hint: use a conditional statement.
  
  * Group each series by school
  
  * Combine the series into a dataframe
  
  * Optional: give the displayed data cleaner formatting

```python
ninth_grade = school_data_complete.loc[school_data_complete['grade'] == '9th'].groupby('school_name')
tenth_grade = school_data_complete.loc[school_data_complete['grade'] == '10th'].groupby('school_name')
eleventh_grade = school_data_complete.loc[school_data_complete['grade'] == '11th'].groupby('school_name')
twelvth_grade = school_data_complete.loc[school_data_complete['grade'] == '12th'].groupby('school_name')

ninth_grade_math = ninth_grade['math_score'].mean()
tenth_grade_math = tenth_grade['math_score'].mean()
eleventh_grade_math = eleventh_grade['math_score'].mean()
twelvth_grade_math = twelvth_grade['math_score'].mean()
```

```python
schools_avg_math = pd.concat([ninth_grade_math,tenth_grade_math,eleventh_grade_math,twelvth_grade_math], axis=1)
schools_avg_math.set_axis(['9th','10th','11th','12th'], axis=1, inplace=True)

schools_avg_math
```

## Reading Score by Grade 

* Perform the same operations as above for reading scores

```python
ninth_grade_reading = ninth_grade['reading_score'].mean()
tenth_grade_reading = tenth_grade['reading_score'].mean()
eleventh_grade_reading = eleventh_grade['reading_score'].mean()
twelvth_grade_reading = twelvth_grade['reading_score'].mean()

schools_avg_reading = pd.concat([ninth_grade_reading,tenth_grade_reading,eleventh_grade_reading,twelvth_grade_reading], axis=1)
schools_avg_reading.set_axis(['9th','10th','11th','12th'], axis=1, inplace=True)

schools_avg_reading
```
